```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
from IPython.display import display
from matplotlib import cm
import plotly.graph_objects as go
plt.rcParams['figure.facecolor'] = 'white'
matplotlib.use('Qt5Agg')
```

```python
path='D:\\ost\\DAP Data\\'
file='0.pkl'
filepath=path+file
df=pd.read_pickle(filepath)
df=df.data[0]
```

```python
display(df.head())
display(df.shape)
display(df.columns)
def get_df_mask(array:np.ndarray)->pd.DataFrame:
    df=pd.DataFrame(array)
    after=df.copy()
    after[after==0.]=np.nan
    after=np.divide(after,after)
    return(after)
```

```python
example_step=1500
mask=get_df_mask(df[('displacements',              'displacement_x')][1500])
plt.matshow(mask)
_1500x=(np.multiply(df[('displacements',              'displacement_x')][example_step],mask))
_1500y=(np.multiply(df[('displacements',              'displacement_y')][example_step],mask))
_1500z=(np.multiply(df[('displacements',              'displacement_z')][example_step],mask))
#display(df[('displacements',              'displacement_x')][example_step])
plt.matshow(_1500x)
plt.colorbar()

plt.matshow(_1500y)
plt.colorbar()

plt.matshow(_1500z)
plt.colorbar()
```

```python
x=range(_1500x.shape[1])
y=range(_1500x.shape[0])
display(x,y)
X,Y=np.meshgrid(x,y)
Z=np.zeros(shape=_1500x.shape)
base=[np.multiply(Y,mask),np.multiply(X,mask),np.multiply(Z,mask)]
data=[np.add(_1500y,base[0]),
                          np.add(_1500x,base[1]),
                          np.add(_1500z,base[2]),]
display(type(data[2]))
plt.matshow(Z)
plt.colorbar()
```

```python
fig = plt.figure(figsize=[20,15])
ax = fig.add_subplot(projection='3d')
#bexis=ax.plot_surface(*base)
axis = ax.plot_wireframe( *data, color='k',linewidth=0.5, antialiased=False,zorder=1)
levels=np.linspace(-0.5, 1.8, 50
                   )
ax.contour(data[0].values,data[1].values,data[2].values, levels,zdir='z', offset=-5.0, cmap=cm.viridis,zorder=0)
ax.plot_surface(base[0],base[1],base[2]+4,cmap=data[2])
ax.view_init(30       ,50)
ax.set_zlim(-5,5)
plt.colorbar(axis)
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
fig = make_subplots(rows=1, cols=1,
                    specs=[[{'is_3d': True}]],
                    subplot_titles=['Color corresponds to z'],
                    )

fig.add_trace(go.Surface(x=np.add(_1500y,base[0]).values,
                         y=np.add(_1500x,base[1]).values,
                         z=np.add(_1500z,base[2]).values,
                         colorscale='viridis'
                         ), 1, 1)
fig.add_trace(go.Surface(x=base[0],y=base[1],z=base[2],
                         colorscale='plasma'
                         ), 1, 1)
fig.update_layout( scene = dict(
        xaxis = dict(nticks=4, range=[0,90],),
                     yaxis = dict(nticks=4, range=[0,90],),
                     zaxis = dict(nticks=4, range=[-10,10],),),
    width=700,
    margin=dict(r=20, l=10, b=10, t=10))
```
