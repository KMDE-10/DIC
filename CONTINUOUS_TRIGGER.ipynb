```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
rd.seed(42,version=2)
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
# for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["txfonts"],
})
```

```python
ratemax=1/1000
def pulse_pause(pulselen,level,ratemax):
    len_pulse=pulselen
    len_gap=rd.randint((len_pulse),(10*len_pulse))
    length=(1/ratemax)*(len_pulse+len_gap)
    pulse=np.full(int(pulselen/ratemax),level)
    #pulse2=np.full(int(pulselen/ratemax/10),level)
    #pulse=np.add(pulse1,pulse2)
    gap=np.zeros(int(len_gap/ratemax),dtype='int')
    full=np.concatenate([pulse,gap])
    #display(pulse.shape,gap.shape,full.shape)
    return full
pulse_pause(1,2,ratemax)
#display(pulse_pause(3,2,ratemax))
print(rd.randint(1,50))
print(rd.randint(1,50))
```

```python
plt.style.use('C:\\Users\\maede\\.matplotlib\\konradstype.mplstyle')
fig,ax=plt.subplots(figsize=(20,10))
ax.set_yticks(range(0,12,2))
start=pulse_pause(1,2,ratemax)
for i in range(80):
    start=np.concatenate([start,pulse_pause(1,2,ratemax)])

ax.plot(start)
ax.set_xlabel(r"$\mathbf{Theo}$")
plt.savefig('hahahah.svg',dpi=500)
```
